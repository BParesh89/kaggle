```python
#lets combine train and test dataset
combined_data  =pd.concat([train.drop('Cover_Type',axis=1),test])
combined_data.shape
```

```python
#lets create a new feature euclidian distance to hydrology
combined_data['Euclidean_Distance_To_Hydrology'] = np.sqrt(np.square(combined_data['Horizontal_Distance_To_Hydrology']) + np.square(combined_data['Vertical_Distance_To_Hydrology']))
combined_data.drop(['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology'],inplace=True,axis=1)
combined_data.head()
```

```python
#lets normalize the features
from sklearn.preprocessing import MinMaxScaler,Normalizer,StandardScaler
#cols_StdSclr = list(combined_data.iloc[:,[2,3,7,8,9]].columns)
#cols_minmax = list(combined_data.iloc[:,[1,4,5,6,10]].columns)
#cols = list(combined_data.columns[1:11])
cols = list(combined_data.iloc[:,[1,2,3,4,5,6,7,8,9,53]].columns)
# print(cols_StdSclr)
# print(cols_Norm)
muffin = list(combined_data.columns[10:-1])
# print(muffin)
# StdSclr = StandardScaler().fit_transform(combined_data[cols_StdSclr])
# StdSclr_form=pd.DataFrame(StdSclr, columns=cols_StdSclr)
# minmax= MinMaxScaler().fit_transform(combined_data[cols_minmax]) 
# minmax_form=pd.DataFrame(minmax,columns=cols_minmax)
StdSclr= StandardScaler().fit_transform(combined_data[cols]) 
StdSclr_form=pd.DataFrame(StdSclr,columns=cols)
```

```python
# norm_form.index
#type(combined_data[muffin])
```

```python
#lets include normalized feature in combined dataset
combined_data1 = pd.concat([StdSclr_form.reset_index(drop=True),combined_data[muffin].reset_index(drop=True)],axis=1)
#combinded_data1 = pd.concat([StdSclr_form.reset_index(drop=True),minmax_form.reset_index(drop=True),combined_data[muffin].reset_index(drop=True)],axis=1)
combined_data1.columns
```

```python
#import classification models
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, learning_curve
```

```python
#seperate train and test dataset
#print(len(train))
X_train = combined_data1[:len(train)]
X_test = combined_data1[len(train):]
y = train['Cover_Type']
kfold = StratifiedKFold(n_splits=10,random_state=0)
```

```python
# Modeling step Test differents algorithms 
random_state = 2
classifiers = []
classifiers.append(SVC(random_state=random_state))
classifiers.append(DecisionTreeClassifier(random_state=random_state))
classifiers.append(AdaBoostClassifier(DecisionTreeClassifier(random_state=random_state),random_state=random_state,learning_rate=0.1))
classifiers.append(RandomForestClassifier(random_state=random_state))
classifiers.append(ExtraTreesClassifier(random_state=random_state))
classifiers.append(GradientBoostingClassifier(random_state=random_state))
classifiers.append(MLPClassifier(random_state=random_state))
classifiers.append(KNeighborsClassifier())
classifiers.append(LogisticRegression(random_state = random_state))
classifiers.append(LinearDiscriminantAnalysis())

cv_results = []
for classifier in classifiers :
    cv_results.append(cross_val_score(classifier, X_train, y, scoring = "accuracy", cv = kfold, n_jobs=4))

cv_means = []
cv_std = []
for cv_result in cv_results:
    cv_means.append(cv_result.mean())
    cv_std.append(cv_result.std())

cv_res = pd.DataFrame({"CrossValMeans":cv_means,"CrossValerrors": cv_std,"Algorithm":["SVC","DecisionTree","AdaBoost",
"RandomForest","ExtraTrees","GradientBoosting","MultipleLayerPerceptron","KNeighboors","LogisticRegression","LinearDiscriminantAnalysis"]})

g = sns.barplot("CrossValMeans","Algorithm",data = cv_res, palette="Set3",orient = "h",**{'xerr':cv_std})
g.set_xlabel("Mean Accuracy")
g = g.set_title("Cross validation scores")
```

```python
rfc = RandomForestClassifier()
rf_param_grid = {"max_depth": [None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "min_samples_leaf": [1, 3, 10],
              "bootstrap": [False],
              "n_estimators" :[300,600],
              "criterion": ["gini"]}


gsRFC = GridSearchCV(rfc,param_grid = rf_param_grid, cv=kfold, scoring="accuracy", n_jobs= 4, verbose = 1)

gsRFC.fit(X_train,y)

RFC_best = gsRFC.best_estimator_

# Best score
gsRFC.best_score_
```

```python
DTC = DecisionTreeClassifier()

adaDTC = AdaBoostClassifier(DTC, random_state=7)

ada_param_grid = {"base_estimator__criterion" : ["gini", "entropy"],
              "base_estimator__splitter" :   ["best", "random"],
              "algorithm" : ["SAMME","SAMME.R"],
              "n_estimators" :[1,5],
              "learning_rate":  [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3,1.5]}

gsadaDTC = GridSearchCV(adaDTC,param_grid = ada_param_grid, cv=kfold, scoring="accuracy", n_jobs= 4, verbose = 1)

gsadaDTC.fit(X_train,y)

ada_best = gsadaDTC.best_estimator_
gsadaDTC.best_score_
```

```python
gdb = GradientBoostingClassifier(random_state=0)
gdb_param_grid = {'loss' : ["deviance"],
              'n_estimators' : [100,200,300],
              'learning_rate': [0.1, 0.05, 0.01],
              'max_depth': [4, 8],
              'min_samples_leaf': [100,150],
              'max_features': [0.3, 0.1] 
              }

gsGBC = GridSearchCV(gdb,param_grid = gdb_param_grid, cv=kfold, scoring="accuracy", n_jobs= 4, verbose = 1)

gsGBC.fit(X_train,y)

GBC_best = gsGBC.best_estimator_

# Best score
gsGBC.best_score_
```

```python
#ExtraTrees 
ExtC = ExtraTreesClassifier()


## Search grid for optimal parameters
ex_param_grid = {"max_depth": [None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "min_samples_leaf": [1, 3, 10],
              "bootstrap": [False],
              "n_estimators" :[100,300],
              "criterion": ["gini"]}


gsExtC = GridSearchCV(ExtC,param_grid = ex_param_grid, cv=kfold, scoring="accuracy", n_jobs= 4, verbose = 1)

gsExtC.fit(X_train,y)

ExtC_best = gsExtC.best_estimator_

# Best score
gsExtC.best_score_
```

Lets check feature importance in tree base algorithms

```python
nrows = ncols = 2
fig, axes = plt.subplots(nrows = nrows, ncols = ncols, sharex="all", figsize=(15,20))

names_classifiers = [("AdaBoosting", ada_best),("ExtraTrees",ExtC_best),("RandomForest",RFC_best),("GradientBoosting",GBC_best)]

nclassifier = 0
for row in range(nrows):
    for col in range(ncols):
        name = names_classifiers[nclassifier][0]
        classifier = names_classifiers[nclassifier][1]
        indices = np.argsort(classifier.feature_importances_)[::-1][:40]
        g = sns.barplot(y=X_train.columns[indices][:40],x = classifier.feature_importances_[indices][:40] , orient='h',ax=axes[row][col])
        g.set_xlabel("Relative importance",fontsize=12)
        g.set_ylabel("Features",fontsize=12)
        g.tick_params(labelsize=9)
        g.set_title(name + " feature importance")
        nclassifier += 1
```

```python
test_Survived_RFC = pd.Series(RFC_best.predict(X_test), name="RFC")
test_Survived_ExtC = pd.Series(ExtC_best.predict(X_test), name="ExtC")
test_Survived_AdaC = pd.Series(ada_best.predict(X_test), name="Ada")
test_Survived_GBC = pd.Series(GBC_best.predict(X_test), name="GBC")


# Concatenate all classifier results
ensemble_results = pd.concat([test_Survived_RFC,test_Survived_ExtC,test_Survived_AdaC,test_Survived_GBC],axis=1)


g= sns.heatmap(ensemble_results.corr(),annot=True)
```

```python
votingC = VotingClassifier(estimators=[('rfc', RFC_best), ('extc', ExtC_best),
('adac',ada_best),('gbc',GBC_best)], voting='soft', n_jobs=4)

votingC = votingC.fit(X_train, y)
```

```python
#predict output
test['Cover_Type'] = votingC.predict(X_test).astype(int)
output = test[['Id','Cover_Type']]
output.to_csv('output.csv',index=False)
```
